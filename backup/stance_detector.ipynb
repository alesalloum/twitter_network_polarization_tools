```python
import json
```

Stance detection is the extraction of a subject's reaction to a claim made by a primary actor. 

```python
# Set the hashtag you want to study
ht = "ilmastonmuutos"
```

```python
# Load the related tweets
with open(ht + "/" + ht + "_structs.json", "r") as read_file:
    data = json.load(read_file)
```

```python
# The number of relevant tweets
print(len(data))
```

```python
# Load the communities
com1, com2 = [], []

with open(ht + "/" + ht + "_community1.txt") as f1:
    lines = f1.readlines()

for line in lines:
    line = line.strip()
    com1.append(line)
    
with open(ht + "/" + ht + "_community2.txt") as f2:
    lines = f2.readlines()

for line in lines:
    line = line.strip()
    com2.append(line)
```

```python
#polarized_com_left = [2628036929910872337, 5927897940583200920, 6741579126382529169, 3010385014743818620, 2914983339049217966]
#polarized_com_right = [8239732675232086996, 2808541593962838809, 1990711970496682409, 128392970747233285, 7763614563692040382]
from random import sample

com1_sample = sample(com1, 5)
com2_sample = sample(com2, 5)


for i in range(len(data)):
    
    if str(data[i]["user"]["id"]) in com1_sample:
        print("COMMUNITY 1\n")
        print("\n")
        print(data[i]["user"]["id"])
        print("\n")
        print(data[i]["text"])
        print("\n")
        
    elif str(data[i]["user"]["id"]) in com2_sample:
        print("COMMUNITY 2\n")
        print("\n")
        print(data[i]["user"]["id"])
        print("\n")
        print(data[i]["text"])
        print("\n")
        
    else:
        continue
```
